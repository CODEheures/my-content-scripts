```python
import pandas as pd

df_meta_data = pd.read_csv("data/articles_metadata.csv", sep=",", header=0, index_col=0)
df_meta_data.reset_index(drop=True, inplace=True)
df_meta_data.head()
```

```python
df_meta_data[df_meta_data.isna()].sum(axis=1).value_counts()
```

```python
df_meta_data.describe()
```

```python
df_embeddings = pd.DataFrame(pd.read_pickle('data/articles_embeddings.pickle'))
df_embeddings.head()
```

```python
df_embeddings[df_embeddings.isna()].sum(axis=1).value_counts()
```

```python
df_articles = pd.concat([df_meta_data, df_embeddings], axis=1)
df_articles.head()
```

```python
import os
df_clicks = None
for f in os.listdir("data/clicks")[0:10]:
    current_df = pd.read_csv(f"data/clicks/{f}", sep=",", header=0)
    df_clicks = pd.concat([df_clicks, current_df], axis=0, ignore_index=True)

len(df_clicks)
```

```python
df_clicks.describe()
```

```python
df = pd.merge(df_clicks, df_articles, how='inner', left_on='click_article_id', right_index=True)
df.head()
```

```python
df.describe().T
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_columns = ['session_start', 'session_size', 'click_timestamp', 'click_environment', 'click_deviceGroup', 'click_os', 'click_country', 'click_region', 'click_referrer_type', 'category_id', 'created_at_ts', 'publisher_id', 'words_count']
scaled = scaler.fit_transform(df[scaled_columns])
df_scaled = df.drop(columns=['session_id'])
df_scaled[scaled_columns] = scaled


article_id = df['click_article_id']
df_scaled = df_scaled.drop(columns=['click_article_id'])
df_scaled.insert(loc=1, column='click_article_id', value=article_id)

df_scaled.describe().round(2)
```
